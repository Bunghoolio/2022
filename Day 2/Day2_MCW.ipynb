```python
df_rps = pd.read_csv('StrategyGuide.txt', header = None, sep = ' ',names = ['Opponent Play','Should Play'])
df_rps2 = pd.read_csv('StrategyGuide.txt', header = None, sep = ' ',names = ['Opponent Play','Round Result'])

print(df_rps)
```

```python
def play_score(value):
	if value == 'rock':
		return 1
	elif value == 'paper':
		return 2
	elif value == 'scissor':
		return 3


df_rps['play_score'] = df_rps['Should Play'].map(play_score)
display(df_rps.head())



```

```python

conditions = [
    ((df_rps['Opponent Play'] == 'rock') & (df_rps['Should Play'] == 'paper')),
    ((df_rps['Opponent Play'] == 'rock') & (df_rps['Should Play'] == 'scissor')),
    ((df_rps['Opponent Play'] =='rock') & (df_rps['Should Play'] == 'rock')),

    ((df_rps['Opponent Play'] == 'paper') & (df_rps['Should Play'] == 'scissor')),
    ((df_rps['Opponent Play'] == 'paper') & (df_rps['Should Play'] == 'rock')),
    ((df_rps['Opponent Play'] =='paper') & (df_rps['Should Play'] == 'paper')),

    ((df_rps['Opponent Play'] == 'scissor') & (df_rps['Should Play'] == 'rock')),
    ((df_rps['Opponent Play'] == 'scissor') & (df_rps['Should Play'] == 'paper')),
    ((df_rps['Opponent Play'] =='scissor') & (df_rps['Should Play'] == 'scissor'))
    ]


values = ['win', 'loss', 'draw','win', 'loss', 'draw','win', 'loss', 'draw']


df_rps['outcome'] = np.select(conditions, values)


print(df_rps)
```

```python
def win_score(value):
    if value == 'win':
        return 6
    if value == 'draw':
        return 3
    if value == 'loss':
        return 0

df_rps['win_score'] = df_rps['outcome'].map(win_score)
```

```python
df_rps['total_score'] = df_rps['play_score'] + df_rps['win_score']
```
