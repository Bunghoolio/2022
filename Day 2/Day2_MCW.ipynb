```python
import numpy as np
import pandas as pd
import math 
```

```python
df_rps = pd.read_csv('StrategyGuide.txt', header = None, sep = ' ',names = ['Opponent Play','Should Play'])


print(df_rps)
```

```python
df_rps = df_rps.replace(['A','B','C','X','Y','Z'], ['rock','paper','scissor','rock','paper','scissor'])
```

```python
for row in df_rps.itertuples(index = False):
    print(row[0],row[1])
    
    
```

```python
def play_score(value):
	if value == 'rock':
		return 1
	elif value == 'paper':
		return 2
	elif value == 'scissor':
		return 3


df_rps['play_score'] = df_rps['Should Play'].map(play_score)
display(df_rps.head())



```

```python

conditions = [
    (df_rps['Opponent Play'] == 'rock' and df_rps['Should Play'] == 'paper'),
    (df_rps['Opponent Play'] == 'rock' and df_rps['Should Play'] == 'scissor'),
    (df_rps['Opponent Play'] =='rock' and df_rps['Should Play'] == 'rock'),

    (df_rps['Opponent Play'] == 'paper' and df_rps['Should Play'] == 'scissor'),
    (df_rps['Opponent Play'] == 'paper' and df_rps['Should Play'] == 'rock'),
    (df_rps['Opponent Play'] =='paper' and df_rps['Should Play'] == 'paper'),

    (df_rps['Opponent Play'] == 'scissor' and df_rps['Should Play'] == 'rock'),
    (df_rps['Opponent Play'] == 'scissor' and df_rps['Should Play'] == 'paper'),
    (df_rps['Opponent Play'] =='scissor' and df_rps['Should Play'] == 'scissor')
    ]


values = ['win', 'loss', 'draw','win', 'loss', 'draw','win', 'loss', 'draw']


df_rps['win_score'] = np.select(conditions, values)


df_rps.head()
```
